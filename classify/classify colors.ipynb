```python
def rgb_classfy(rgb):
    r = rgb[0]; g = rgb[1]; b = rgb[2]
    
    if ((r/g) >= 0.95 and (r/g) <= 1.05) and ((g/b) >= 0.95 and (g/b) <= 1.05) and ((r/g) >= 0.95 and (r/g) <= 1.05):
        return 'mono'
    elif r < 51 and g < 51 and b < 51:
        return 'mono'
    else:
        if r > b:
            return 'red'
        else:
            return 'blue'
```

```python
reds = dict()
blues = dict()
monos = dict()

for name in pathes[:10]:
    image_name = 'C:/Users/rlath/dataitgirls/color_on_seoul_cafe/cafe_image/' + name
    image = cv2.imread(image_name, cv2.IMREAD_COLOR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    image_list = [list(lst) for img in image for lst in img]

    rb = [rgb_classfy(rgb) for rgb in image_list]

    monos[name] = rb.count('mono')
    reds[name] = rb.count('red')
    blues[name] = rb.count('blue')
```
